## 04 実際にコードを実行してコードを修正してみよう

```python
# 必要なライブラリのimport ※selenium,time,os,datetime,pandas
from selenium import webdriver
import time
import os
import datetime
import pandas as pd

def main():
    '''
    メインの実行部分
    '''
    # 何年前からのデータを取得するのかを入力
    year = 2015
    # 今年の西暦を取得
    this_year = datetime.date.today().year

    # 銘柄コードの入力
    stock_code = 6758
    # Seleniumでダウンロードするデフォルトディレクトリ
    download_directory = r"C:\Work\Python\20201109_Daytry_Python\Day29\prophet\stockdata_test"
    # CSVのダウンロードの実行
    download(stock_code, year, this_year, download_directory)
    # ダンロードしたデータフレームの結合
    df = concat_df(stock_code, year, this_year, download_directory)
    # 結合したデータフレームをCSVファイルで出力
    df.to_csv(f"{download_directory}/{stock_code}_{year}_{this_year}.csv")
    # 完成したデータフレームの出力
    print(df)

def download(stock_code, year, this_year, download_directory):
    '''
    特定の銘柄のCSVデータを各年ごとにダウンロード（デフォルトディレクトリの指定）
    '''
    options = webdriver.ChromeOptions()  # クロームオプションの設定
    prefs = {"download.default_directory" : download_directory}  # デフォルトのダウンロードディレクトリの指定
    options.add_experimental_option("prefs",prefs)
    driver = webdriver.Chrome(options=options, executable_path=r"C:\Work\Python\20201109_Daytry_Python\Day5\google_weak_keyword_search\chromedriver.exe")
  # オプションを指定してクロームドライバーの起動
    driver.implicitly_wait(30)  # 要素が見つからなかった時の待ち時間を指定
    
    # 各年のCSVデータのダウンロードをfor文で繰り返す
    for y in range(year, this_year + 1):
        driver.get(f"https://kabuoji3.com/stock/{stock_code}/")
        # ファイルがすでに存在している場合はcontinueで次のyへ行く
        if os.path.isfile(f"{download_directory}/{stock_code}_{y}.csv"):
            # 3秒待機
            time.sleep(3)
            continue
        # seleniumでy年のテキストをクリック
        driver.find_element_by_link_text(str(y)).click()
        # 3秒待機
        time.sleep(3)
        # クリックでCSVのダウンロード
        driver.find_element_by_name("csv").click()
        driver.find_element_by_name("csv").click()
        # 3秒待機
        time.sleep(3)
    # ブラウザを閉じて終了する
    driver.quit()
    
def concat_df(stock_code, year, this_year, download_directory):
    '''
    ダウンロードしたCSVをデータフレームとして結合
    '''
    # 結合するデータフレームを入れる空のリストを用意
    concat_list = []
    
    # データフレームの結合
    for y in range(year, this_year + 1):
        # データフレームの読み込み
        df = pd.read_csv(f"{download_directory}/{stock_code}_{y}.csv", encoding = 'shift-jis')
        # 日付行の削除
        df.drop(index='日付', inplace=True)
        # pd.concatで結合するためにリストにデータフレームを入れていく
        concat_list.append(df)
    # pd.concatでデータフレームの結合
    df = pd.concat(concat_list)
    return df

if __name__ == "__main__":
    main()
```

## 01 prophetを使って株価予測のコードを書いていこう

```python
import pandas as pd
from fbprophet import Prophet

# データフレームの読み込み
df = pd.read_csv(r'C:\Work\Python\20201109_Daytry_Python\Day29\prophet\stockdata_test\6758_2013_2020.csv')
# データフレームの出力
df
```

```python
# データフレームの出力
df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']

df
```

```python
# カラム名をdsとyに変更
df = df.rename(columns={'Date':'ds', 'Adj Close':'y'})
df
```

```python
# Prophetのインスタンス化
m = Prophet()
m
```

```python
# fitメソッドでデータを学習させる
m.fit(df)
```

```python
# make_future_dataframeメソッドで予測分の未来のデータフレームの作成
future = m.make_future_dataframe(periods=250)
future
```

```python
future.plot()
```

```python
future['ds'].dt.weekday
```

```python
#株の取り引きが月曜から金曜日までしかやっていない

# 月曜=0,火曜=1,水曜=2,木曜=3,金曜=4,土曜=5,日曜=6
future = future[future['ds'].dt.weekday < 5]
```

```python
future
```

```python
# 予測する
forecast = m.predict(future)
forecast
```

```python
# 予測結果の描画 jupyter notebook
import matplotlib.pyplot as plt
%matplotlib inline
fig1 = m.plot(forecast)
fig1

plt.show()
```

```python
# トレンド、周期性の描画
fig2 = m.plot_components(forecast)
fig2
```

```python
# 必要なライブラリのimport ※selenium,time,os,datetime,pandas
from selenium import webdriver
import time
import os
import datetime
import pandas as pd

def main():
    '''
    メインの実行部分
    '''
    search_keyword_list = ['重工業','金融']
    
    # 何年前からのデータを取得するのかを入力
    year = 2015
    # 今年の西暦を取得
    this_year = datetime.date.today().year

    df_serch_code = make_df_stock_code(search_keyword_list)
    
    for index, stock_code in enumerate(df_serch_code['コード']): # 銘柄コードの入力
        # 銘柄名の取得
        company = df_serch_code['銘柄名'].iloc[index].replace('\u3000','_')
        # Seleniumでダウンロードするデフォルトディレクトリ
        download_directory = r"C:\Work\Python\20201109_Daytry_Python\Day29\prophet\stockdata_list"
    
        # CSVのダウンロードの実行
        download(stock_code, year, this_year, download_directory)
        # ダンロードしたデータフレームの結合
        df = concat_df(stock_code, year, this_year, download_directory)
        # 結合したデータフレームをCSVファイルで出力
        df.to_csv(f"{download_directory}/{company}_{stock_code}_{year}_{this_year}.csv")
        # 完成したデータフレームの出力
        
    print(df)

def download(stock_code, year, this_year, download_directory):
    '''
    特定の銘柄のCSVデータを各年ごとにダウンロード（デフォルトディレクトリの指定）
    '''
    options = webdriver.ChromeOptions()  # クロームオプションの設定
    prefs = {"download.default_directory" : download_directory}  # デフォルトのダウンロードディレクトリの指定
    options.add_experimental_option("prefs",prefs)
    driver = webdriver.Chrome(options=options, executable_path=r"C:\Work\Python\20201109_Daytry_Python\Day5\google_weak_keyword_search\chromedriver.exe")
  # オプションを指定してクロームドライバーの起動
    driver.implicitly_wait(30)  # 要素が見つからなかった時の待ち時間を指定
    
    # 各年のCSVデータのダウンロードをfor文で繰り返す
    for y in range(year, this_year + 1):
        driver.get(f"https://kabuoji3.com/stock/{stock_code}/")
        # ファイルがすでに存在している場合はcontinueで次のyへ行く
        if os.path.isfile(f"{download_directory}/{stock_code}_{y}.csv"):
            # 3秒待機
            time.sleep(3)
            continue
        # seleniumでy年のテキストをクリック
        try:
            driver.find_element_by_link_text(str(y)).click()
            # 3秒待機
            time.sleep(3)
            # クリックでCSVのダウンロード
            driver.find_element_by_name("csv").click()
            driver.find_element_by_name("csv").click()
            ystart = y
        except Exception as e:
            print(e)
            print(f'{y}年はありません。')
#             print(f'{company}_{stock_code}の{y}年は存在しませんでした。')        

        # 3秒待機
        time.sleep(3)
    # ブラウザを閉じて終了する
    driver.quit()
    
def concat_df(stock_code, year, this_year, download_directory):
    '''
    ダウンロードしたCSVをデータフレームとして結合
    '''
    # 結合するデータフレームを入れる空のリストを用意
    concat_list = []
    
    # データフレームの結合
    for y in range(year, this_year + 1):
        try:
            # データフレームの読み込み
            df = pd.read_csv(f"{download_directory}/{stock_code}_{y}.csv", encoding = 'shift-jis')
            # 日付行の削除
            df.drop(index='日付', inplace=True)
            # pd.concatで結合するためにリストにデータフレームを入れていく
            concat_list.append(df)
        except Exception as e:
            print(e)

    # pd.concatでデータフレームの結合
    df = pd.concat(concat_list)
    return df

def make_df_stock_code(search_keyword_list):
    
    # 株価コードの読み込み
    stock_code_list = pd.read_excel(r'./prophet/data_j.xls')
    
    
    # 空のデータフレームを作成
    df_serch = pd.DataFrame()

    for keyword in search_keyword_list:
        df_serch_ = stock_code_list[(stock_code_list['銘柄名'].str.contains(keyword))]
        df_serch = pd.concat([df_serch, df_serch_],axis=0)
    
    df_serch_code = df_serch[['コード','銘柄名']]
    return df_serch_code

if __name__ == "__main__":
    main()
```
